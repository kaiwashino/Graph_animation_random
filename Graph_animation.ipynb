```python
%matplotlib notebook
 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
import random
```

```python
fig = plt.figure()
 
xlim = [0,100]
X, Y = [], []
Normal_time = 5000
Change_time = 5000
Change_rate = 2
range1 = 10
range2 = 20
cnt = 1
def plot(data):
    plt.cla()    # å‰ã®ã‚°ãƒ©ãƒ•ã‚’å‰Šé™¤
    cnt = 1
    if cnt < Normal_time:
        Y.append(random.randrange(range1,range2))   # ãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ
        X.append(len(Y))
    elif (Normal_time <= cnt) and (cnt < Normal_time + Change_time):
        Y.append(random.randrange(range1+50,range2+50))   # ãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ
        X.append(len(Y))
    else :
        Y.append(random.randrange(10,20))   # ãƒ‡ãƒ¼ã‚¿ã‚’ä½œæˆ
        X.append(len(Y))
    if len(X) > 100:            # æç”»ç¯„å›²ã‚’æ›´æ–°
        xlim[0]+=1
        xlim[1]+=1
    plt.plot(X, Y)              # æ¬¡ã®ã‚°ãƒ©ãƒ•ã‚’ä½œæˆ
    plt.title("Saliva volume")
    plt.ylim(0,150)
    plt.xlim(xlim[0],xlim[1])
    
    # 10msã”ã¨ã«ploté–¢æ•°ã‚’å‘¼ã³å‡ºã—ã¦ã‚¢ãƒ‹ãƒ¡ãƒ¼ã‚·ãƒ§ãƒ³ã‚’ä½œæˆ
ani = animation.FuncAnimation(fig, plot, interval=10, blit=True)
#ani.save('sample2.gif', writer='imagemagick')
plt.show()
```
